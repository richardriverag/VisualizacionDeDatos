#### Ejemplo básico de funcionamiento

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de ejemplo
data = {
    'Categoria': ['A', 'B', 'C', 'D'],
    'Valor1': [10, 20, 15, 25],
    'Valor2': [12, 18, 20, 10],
    'Valor3': [8, 22, 10, 30]
}
df = pd.DataFrame(data).set_index('Categoria')

# Definición del gráfico
ax = df.plot(kind='bar', stacked=True, title='Gráfico de Barras Apiladas', rot=0)

# Añadimos una leyenda para los colores
ax.legend(title='Categorías')
plt.show()
```

####  Ventajas  y Desventajas de los Gráficos de Barras Apiladas

| Ventajas                                  |  Desventajas                                 |
|---------------------------------------------|--------------------------------------------|
| Visión dual (Total + Composición)           | Dificultad para comparar segmentos internos|
| Eficiencia de espacio                       | Sobrecarga visual (el "efecto arcoíris")   |
|Identificación rápida de tendencias generales|Oculta la tendencia de los subgrupos        |

* **Tipos de Variables Requeridas**
  * Dos variables cualitativas (categóricas o temporales): 
    * *Variable principal:* Define las barras completas en el eje base (ej. Meses, Países, Departamentos).
    * *Variable secundaria (subgrupos):* Define los segmentos de color dentro de cada barra (ej. Canal de tráfico, Género, Tipo de producto).
  * Una variable cuantitativa (numérica): Define la longitud o altura de cada segmento (ej. Ventas en dólares, Cantidad de personas). 

* **Limitaciones de Tamaño y Visualización**
  * **Tamaño Mínimo:** Al menos 2 categorías principales y 2 subgrupos. Si tienes menos, un gráfico de barras simple o de pastel es suficiente.
  * **Límite Máximo en Eje Base (Barras):** Entre 10 y 15 barras. Más allá de eso, las etiquetas se amontonan y el gráfico pierde legibilidad (a menos que se use un gráfico horizontal).
  * **Límite Máximo en Apilamiento (Segmentos):** Máximo 4 a 5 subgrupos por barra. Exceder esto genera sobrecarga cognitiva y dificultad para distinguir los colores.

## 1.3 Realización práctica de una visualización simple
###  Sobre el Dataset: StatCounter Global Stats

Este conjunto de datos rastrea la cuota de mercado global de sistemas operativos <a href="#fn2" id="ref2">[2]</a>. 
Los valores representan el porcentaje de uso de plataformas como Windows, Android e iOS desde el año 2009 al 2026.

<p id="fn2">[2] <a href="[url](https://gs.statcounter.com/os-market-share#yearly-2009-2026)">Fuente: StatCounter Global Stats - OS Market Share</a><a href="#ref2">
</a></p>


```python
import pandas as pd
import matplotlib.pyplot as plt
import io

# Cargamos los datos 
df = pd.read_csv('data/os_combined-ww-yearly-2009-2026.csv', index_col='Date')

# Limpiamos y agrupamos los datos (Para evitar el efecto arcoíris)
# Seleccionamos solo los líderes del mercado para las capas principales
top_os = ['Windows', 'Android', 'iOS', 'OS X', 'Linux']

# Sumamos todas las demás columnas en una nueva categoría llamada 'Otros'
df['Otros'] = df.drop(columns=top_os).sum(axis=1)
#print(df)

# Como tenemos 18 columnas esto podría saturar el gráfico, vamos a tomar solo los años pares
df = df[df.index % 2 == 0]
# Filtramos nuestro DataFrame para que solo tenga las columnas que queremos graficar
df_plot = df[top_os + ['Otros']]


# Configuramos el tamaño de la figura
fig, ax = plt.subplots(figsize=(10, 6))

# Para que sea un gráfico de barras apiladas en pandas usamos los argumentos kind=bar y stacked=True
df_plot.plot(kind='bar', stacked=True, ax=ax, colormap='Set2', width=0.75, edgecolor='white')

# Personalizamos el gráfico para la presentación
ax.set_title('Evolución de la Cuota de Mercado Global de Sistemas Operativos', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Año', fontsize=12)
ax.set_ylabel('Cuota de Mercado (%)', fontsize=12)
plt.xticks(rotation=0)
ax.legend(title='Sistema Operativo', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()

# Mostramos el gráfico
plt.show()
```

## 1.4 Comentario del gráfico
Este gráfico nos muestra claramente cómo Windows dominaba con más del 90% en 2010 y cómo la barra de Android creció rápidamente año tras año, empujando los demás segmentos.

# 2. Waffle Chart
### 2.1 Definición General
* **Otros nombres:** Gráfico de Cuadrícula (*Gridplot*), Gráfico Circular Cuadrado (*Square Pie Chart*), Gráfico de Unidades.
* **Autor / Origen:** No tiene un inventor histórico único (a diferencia del gráfico de pastel de William Playfair). Evolucionó de forma orgánica en el periodismo de datos moderno y la creación de *dashboards* corporativos como una solución geométrica a las fallas de percepción humana con los ángulos.
* **Descripción** Es una cuadrícula (tradicionalmente de 10x10) donde cada celda representa una unidad o un porcentaje del total (ej. 1 celda = 1%). Las celdas se colorean según la categoría a la que pertenecen para formar un todo.

* **Aplicaciones más comunes:**
    * Mostrar el progreso hacia un objetivo o meta (ej. % de presupuesto gastado).
    * Visualizar la relación de partes de un todo de forma rápida.
    * Resaltar un KPI (Indicador Clave de Rendimiento) principal en infografías.

#### Ejemplo básico de funcionamiento

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle

data = {"Gatos": 50, "Perros": 40, "Conejos": 10}

fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    title={"label": "Popularidad de mascotas", "loc": "center"},
    legend={"loc": "lower left", "bbox_to_anchor": (0, -0.4), "ncol": len(data)}
)

plt.show()
```

####  Ventajas  y Desventajas del Waffle Chart
| Ventajas                                  |  Desventajas                                 |
|---------------------------------------------|--------------------------------------------|
| Nuestro cerebro procesa áreas cuadradas y cuenta unidades con mucha más precisión que los ángulos o rebanadas de un círculo| Ocupa bastante espacio en pantalla|
| No distorsiona visualmente las proporciones pequeñas y es estéticamente muy limpio| Si no se usa una cuadrícula estandarizada (como 10x10), pierde su cualidad de lectura rápida.   |

## 2.2 Tipos y Estructuras de datos

* **Tipo y Estructura de Datos:**
    * **Datos adecuados:** 
      * Una variable cualitativa (categorías) y una variable cuantitativa (porcentajes o valores absolutos).
    * **Requisito específico:** 
      * La suma de los valores cuantitativos debe representar un total lógico (generalmente el 100%).
    * **Limitaciones:** 
      * Se recomienda **estrictamente no usar más de 5 a 6 categorías**. Más allá de eso, la cuadrícula se vuelve un mosaico difícil de leer.

## 2.3 Realización práctica de una visualización simple
###  Sobre el Dataset: StatCounter Global Stats

Este conjunto de datos rastrea la cuota de mercado global de sistemas operativos <a href="#fn2" id="ref2">[2]</a>. 
Los valores representan el porcentaje de uso de plataformas como Windows, Android e iOS desde el año 2009 al 2026.
Para este gráfico solo utilizamos los valores del año 2026.

<p id="fn2">[2] <a href="[url](https://gs.statcounter.com/os-market-share#yearly-2009-2026)">Fuente: StatCounter Global Stats - OS Market Share</a><a href="#ref2">
</a></p>


```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle

# Cargamos los datos
df = pd.read_csv('data/os_combined-ww-yearly-2009-2026.csv', index_col='Date')
top_os = ['Windows', 'Android', 'iOS', 'OS X', 'Linux']

# Preparamos los datos y redondeamos
data_2026 = df.loc[2026, top_os].copy()
dict_2026 = data_2026.round(0).astype(int).to_dict()
dict_2026['Otros'] = 100 - sum(dict_2026.values())

# Definimos el gráfico de Waffle
fig = plt.figure(
    FigureClass=Waffle,
    rows=10, 
    values=dict_2026,
    # Usamos los colores de las marcas para que sea intuitivo
    colors=['#00Bcf2', '#3DDC84', '#555555', '#A3AAAE', "#f80808", '#d3d3d3'],
    title={'label': 'Cuota de Mercado Global de Sistemas Operativos (2026)', 'loc': 'center', 'fontsize': 16, 'fontweight': 'bold','pad': 20},
    labels=[f"{k} ({v}%)" for k, v in dict_2026.items()],
    legend={'loc': 'upper center', 'bbox_to_anchor': (0.5, -0.05), 'ncol': 5, 'frameon': False, 'fontsize': 12},
    figsize=(9, 6)
)

plt.show()
```

## 2.4 Comentario del gráfico
El gráafico ilustra la fragmentación actual del mercado: Android liderando (36.62%), seguido por Windows (30.43%) y el resto repartido.


# 2. Tile grid Map
## 2.1 Definición General
* **Otros nombres:** Mapa de mosaicos, Cartograma de cuadrados/hexágonos (*Square/Hex grid map*), Mapa de celdas iguales.
* **Autor / Origen:** Es una evolución moderna de los *Cartogramas* (popularizados por el geógrafo Danny Dorling en los 90s). Su formato de cuadrícula estricta se popularizó en la última década gracias a medios de periodismo de datos como *The New York Times* y *Bloomberg* para reportar resultados electorales de EE. UU.
* **Descripción:** Es una abstracción de un mapa geográfico donde cada región (estado, provincia o país) está representada por una figura del mismo tamaño (generalmente un cuadrado o un hexágono). Estas figuras se acomodan en una cuadrícula para imitar, de forma aproximada, la geografía real del territorio.

* **Aplicaciones más comunes:**
    * Resultados electorales donde cada región tiene un peso similar o idéntico.
    * Comparativas de métricas socioeconómicas por estado/provincia donde no queremos que el tamaño físico del territorio sesgue la percepción.
    * *Dashboards* interactivos de cobertura nacional.

#### Ejemplo básico de funcionamiento

```python
import matplotlib.patches as patches
import matplotlib.cm as cm
import matplotlib.colors as mcolors
states = pd.read_csv('data/us_tile_grid.xls')
fig, axs = plt.subplots(1, figsize=(12, 6), dpi=300)
plt.xlim(0, 12)
plt.ylim(0, 8)
plt.axis('off')
axs.set_aspect('equal')
for _, state in states.iterrows():
    box = patches.Rectangle((state.column,state.row),1,1,linewidth=1,edgecolor='white',facecolor='lightgrey')
    plt.text(state.column + 0.5, state.row + 0.5, state.code, horizontalalignment='center', verticalalignment='center', color='dimgrey')
    axs.add_patch(box)

plt.show()
```

####  Ventajas  y Desventajas del Waffle Chart
| Ventajas                                  |  Desventajas                                 |
|---------------------------------------------|--------------------------------------------|
| Elimina el "sesgo de área" (el tamaño del territorio no dicta la importancia visual)| Sacrifica la precisión geográfica y topológica (algunas regiones pueden no tocar a sus vecinos reales)|
| Hace que regiones geográficamente minúsculas sean tan visibles como las gigantes| Requiere que el lector ya esté familiarizado con la geografía general del lugar para entender la abstracción   |
|Es estéticamente muy limpio y moderno| Puede ocupar mucho espacio en pantalla|


## 3.2 Tipos y Estructuras de datos

* **Variables requeridas:**
    * **Cualitativa (Identificador):** El nombre o código de la región (ej. "NY", "TX", "California").
    * **Cuantitativa o Cualitativa (Valor a colorear):** La métrica que definirá el color de la celda (ej. Tasa de desempleo, Partido político ganador).
    * **Requisito específico estructural (Coordenadas geográficas abstractas):** Se necesitan dos columnas extra que definan la posición `X` (columna) e `Y` (fila) de cada región dentro de la cuadrícula matemática.
* **Limitaciones de conjunto de datos:**
    * **Tamaño Mínimo:** Alrededor de 15 a 20 regiones. Si son menos, al acomodarlas no parecerán un mapa, solo cuadros sueltos.
    * **Tamaño Máximo:** Lo ideal es usarlo para divisiones de primer nivel (ej. los 50 estados de EE. UU., los 32 estados de México, los países de Europa). Superar los 100-200 elementos (como intentar graficar todos los municipios/condados) convierte al gráfico en un mapa de píxeles difícil de leer, perdiendo el propósito de la cuadrícula.

## 3.3 Realización práctica de una visualización simple
###  Sobre el Dataset: StatCounter Global Stats

Para este gráfico seguimos con StatCounter pero hemos utilizado otro dataset que nos muestra el Sistema Operativo más utilizado por país <a href="#fn3" id="ref3">[3]</a>.
Además, para poder realizar el gráfico necesitamos las coordenadas de los países, para esto hemos descargado las coordenadas para este tipo de mapas utilizado un repositorio de GitHub<a href="#fn4" id="ref4">[4]</a>.

<p id="fn3">[3] <a href="[url](https://gs.statcounter.com/os-market-share#monthly-202602-202602-map)">Fuente: StatCounter Global Stats - OS Market Share</a><a href="#ref3">
</a></p>
<p id="fn4">[4] <a href="[url](https://github.com/mustafasaifee42/Tile-Grid-Map)">Fuente: StatCounter Global Stats - OS Market Share</a><a href="#ref3">
</a></p>

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Cargamos y limpiamos los datos del CSV (Cuotas de Mercado)
df_csv = pd.read_csv('data/os_combined-ww-monthly-202602-202602-map.csv', index_col='Continent')
# Limpiamos espacios en blanco al inicio y al final de los nombres
df_csv.index = df_csv.index.str.strip() 

# Cargamos las coordenadas geográficas del JSON
with open('data/Tile-Grid-Map-Cleaned.json', 'r', encoding='utf-8') as f:
    grid_data = json.load(f)

# Configuramos parámetros y colores
os_columns = ['Android', 'Windows', 'iOS', 'OS X', 'Linux']
colors = {'Windows': '#00Bcf2', 'Android': '#3DDC84', 'iOS': '#555555', 'OS X': '#A3AAAE', 'Linux': '#F5A70A'}

# Preparamos un espacio grande para que quepan todos los países
fig, ax = plt.subplots(figsize=(16, 9))

# Procesamos y dibujamos cada país del JSON
for country_info in grid_data:
    country_name = country_info['name']
    
    # Manejo de un par de excepciones comunes de nombres para ilustrar un ejemplo de no usar un estándar para los nombres
    if country_name == "Viet Nam": country_name = "Vietnam"
    if country_name == "United States": country_name = "United States of America"
    
    x, y = country_info['coordinates']
    alpha3 = country_info['alpha-3'] # Usamos el código de 3 letras (ej. USA, BRA)
    
    # Color por defecto (Gris claro) si no encontramos el país en el CSV
    color = '#e0e0e0' 
    text_color = 'black'
    
    # Si el país está en nuestro CSV, buscamos el SO dominante
    if country_name in df_csv.index:
        # Extraemos la fila de ese país y calculamos el máximo
        os_data = df_csv.loc[country_name, os_columns]
        dominant_os = os_data.idxmax()
        
        # Asignamos el color corporativo
        color = colors.get(dominant_os, '#d3d3d3')
        text_color = 'white' # Texto blanco para contraste sobre colores oscuros
        
    # Dibujamos el cuadrado (Tile) en la coordenada (x, y)
    rect = patches.Rectangle((x, y), 0.9, 0.9, facecolor=color, edgecolor='white', linewidth=0.5)
    ax.add_patch(rect)
    
    # Añadimos el código del país en el centro del cuadrado
    # Reducimos el tamaño de fuente (fontsize=6) porque el mundo tiene muchos cuadritos
    ax.text(x + 0.45, y + 0.45, alpha3, color=text_color, 
            weight='bold', ha='center', va='center', fontsize=6)

# Los sistemas de coordenadas web (como nuestro JSON) suelen tener el 0,0 arriba a la izquierda.
# Matplotlib lo tiene abajo a la izquierda. Invertimos el eje Y para que el norte quede arriba.
ax.invert_yaxis()

# Ajustamos los límites de la gráfica para enmarcar todos los cuadritos
ax.autoscale_view()
ax.set_aspect('equal') # Mantiene los cuadrados perfectos
ax.axis('off') # Ocultamos los ejes numéricos

# Creamos una leyenda personalizada
legend_elements = [patches.Patch(facecolor=c, label=os) for os, c in colors.items()]
legend_elements.append(patches.Patch(facecolor='#e0e0e0', label='Sin Datos / No Coincide'))
ax.legend(handles=legend_elements, loc='lower center', bbox_to_anchor=(0.5, -0.1), 
          ncol=6, frameon=False, fontsize=11)
plt.title('Mapa de Cuadrícula Mundial: SO Dominante (Feb 2026)', fontsize=18, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

## 3.4 Comentario del gráfico
Países enormes geográficamente como Rusia o Canadá tienen exactamente el mismo impacto visual (1 cuadrito) que territorios diminutos pero altamente tecnológicos como Singapur, permitiendo un análisis imparcial del mercado
